### Data Visualization with Matplotlib and Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Let's see how a name like "Daniel" as trended over the years using Seaborn
selected_name = "Daniel"
name_data = us_babies[us_babies["Name"] == selected_name]
name_data.sort_values("Year")
sns.lineplot(data=name_data, x="Year", y="Count")
```

```python
# Let's do the same with matplotlib
plt.plot(name_data["Year"], name_data["Count"])
plt.title(f"How the name '{selected_name}' has trended over the years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
# Let's get the five most popular names between 1880 and 2017 and see the relationship between them
top5_allTime = us_babies.sort_values("Count", ascending=False).head(15)
top5_allTime
```

##### From the above data, we can see that Linda, Micheal, Robert, James and John standout. Let's span them across the years and see their highs and lows

```python
names = ["Linda", "Micheal","Robert", "James", "John"]

for name in names:
    data = us_babies[us_babies["Name"] == name].sort_values("Year")
    plt.plot(data["Year"], data["Count"], label=name)

plt.title("Top 5 names in the US from 1880 to 2017")
plt.xlabel("Year")
plt.ylabel("Count")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Total births per year
births_per_year = us_babies.groupby("Year", as_index=False)["Count"].sum()
births_per_year = births_per_year.rename(columns={"Count":"Total_Births"})
plt.plot(births_per_year["Year"], births_per_year["Total_Births"])
plt.title("Total Births in the US per year from 1880 to 2017")
plt.xlabel("Year")
plt.ylabel("Number of Births")
plt.grid(True)
plt.show()
```

```python
# Most stable names
stable_names = tuple(us_babies.groupby("Name")["Year"].nunique().pipe(lambda x: x[x == us_babies["Year"].nunique()]).index)
print("Total number of names that appeared every year: " + str(len(stable_names)))
```
